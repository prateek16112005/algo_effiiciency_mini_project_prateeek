```python
#bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print(bubble_sort([12, 11, 13, 5, 6]))
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid] 
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

    return arr
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
print(selection_sort([12, 11, 13, 5, 6]))
```

```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

insertion_sort([12, 11, 13, 5, 6])
```

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
import matplotlib.pyplot as plt
import time

def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

def fib_iterative(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a

ns = list(range(1, 31))
recursive_times = []
iterative_times = []

for n in ns:
    start = time.perf_counter()
    fib_recursive(n)
    recursive_times.append(time.perf_counter() - start)

    start = time.perf_counter()
    fib_iterative(n)
    iterative_times.append(time.perf_counter() - start)

plt.figure(figsize=(5,4))
plt.plot(ns, recursive_times, label='Recursive')
plt.plot(ns, iterative_times, label='Iterative')
plt.xlabel('n (Fibonacci Index)')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time vs. Input Size for Fibonacci Implementations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import time
import random


ns = list(range(100, 1000, 300)) 
bubble_sort_times = []
insertion_sort_times = []
selection_sort_times = []
merge_sort_times = []
quick_sort_times = []
binary_search_times = []

for n in ns:
    arr = random.sample(range(n*10), n)  # Create random array of size n

    # Time bubble sort
    arr_copy = arr.copy()
    start = time.perf_counter()
    bubble_sort(arr_copy)
    bubble_sort_times.append(time.perf_counter() - start)

    # Time insertion sort
    arr_copy = arr.copy()
    start = time.perf_counter()
    insertion_sort(arr_copy)
    insertion_sort_times.append(time.perf_counter() - start)

    # Time selection sort
    arr_copy = arr.copy()
    start = time.perf_counter()
    selection_sort(arr_copy)
    selection_sort_times.append(time.perf_counter() - start)
    
    
    arr_copy = arr.copy()
    start = time.perf_counter()
    merge_sort(arr_copy)
    merge_sort_times.append(time.perf_counter() - start)

    # Time quick sort
    arr_copy = arr.copy()
    start = time.perf_counter()
    quick_sort(arr_copy)
    quick_sort_times.append(time.perf_counter() - start)
    
    # Time binary search on a sorted array
    arr_copy = sorted(arr.copy())
    target = arr_copy[n // 2]  # Search for the middle element
    start = time.perf_counter()
    binary_search(arr_copy, target)
    binary_search_times.append(time.perf_counter() - start)
# Plotting the results
plt.figure(figsize=(5,4))
plt.plot(ns, bubble_sort_times, label='Bubble Sort')
plt.plot(ns, insertion_sort_times, label='Insertion Sort')
plt.plot(ns, selection_sort_times, label='Selection Sort')
plt.plot(ns, merge_sort_times, label='Merge Sort')
plt.plot(ns, quick_sort_times, label='Quick Sort')
plt.plot(ns, binary_search_times, label='Binary Search')
plt.xlabel('Array Size (n)')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Sorting Algorithms')
plt.legend()
plt.grid(True)
plt.show()
```
